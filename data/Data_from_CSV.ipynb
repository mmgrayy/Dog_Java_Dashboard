```python
#import dependencies
import os
import csv
import pandas as pd 
```

```python
# File to Load
file_to_load = "Resources/dog_breed_characteristics.csv"
# Read Purchasing File and store into Pandas data frame
dog_characteristics = pd.read_csv(file_to_load, encoding="utf-8")
dog_characteristics
```

```python
 # Remove unecessary columns from the DataFrame and save the new DataFrame
    # Only keep: "BreedName", "Group1", "Temperment", "AvgPupPrice", "PopularityUS2017"

dogs_cleaned_df = dog_characteristics[["BreedName", "Group1", "Temperment", "AvgPupPrice",
                       "PopularityUS2017"]]

dogs_cleaned_df.head()
```

```python
# Push the remade DataFrame to a new CSV file
dogs_cleaned_df.to_csv("Output/dog_breed_characteristics_clean.csv",
                  encoding="utf-8", index=False, header=True)
```
